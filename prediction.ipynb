```python
from tensorflow.keras.models import load_model
import pickle

### load the trained model,scaler pickle,onehot
model = load_model('model(1).h5')

## load the encoder and scaler

with open('onehot_encoder_geo(3).pkl','rb') as file:
  onehot_encoder_geo=pickle.load(file)

with open('label_encoder_gender(3).pkl','rb') as file:
  label_encoder_gender=pickle.load(file)

with open('scaler(3).pkl', 'rb') as file:
  scaler = pickle.load(file)
```

```python
# Example  input data
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000

}
```

```python
import pandas as pd
# one -hot encode 'Geography'
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
## encode categorical variables
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
###concatination one hot encoded
# Drop any existing one-hot encoded geography columns to prevent duplicates on re-execution
geo_cols_to_drop = [col for col in input_df.columns if col.startswith('Geography_')]
input_df = input_df.drop(columns=geo_cols_to_drop, errors='ignore')

# Drop the original 'Geography' column and concatenate the new one-hot encoded columns
input_df=pd.concat([input_df.drop("Geography", axis=1, errors='ignore'),geo_encoded_df],axis=1)
input_df
```

```python
## scaling the data
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
### prediction of the churn
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
  print('The customer is likely to churn.')
else:
  print('The customer is not likely to churn.')
```
